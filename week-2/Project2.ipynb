```python
import numpy as np
import math

print("Equation Analyzer")
print("1 - Quadratic (Ax^2 + Bx + C = 0)")
print("2 - Cubic (Ax^3 + Bx^2 + Cx + D = 0)")
print("3 - Quartic (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

choice = int(input("Choose the equation type (1/2/3): "))

# Quadratic
if choice == 1:
    print("Quadratic Equation")

    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))

    discriminant = b**2 - 4*a*c

    if discriminant >= 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
    else:
        root1 = (-b + complex(0, math.sqrt(-discriminant))) / (2*a)
        root2 = (-b - complex(0, math.sqrt(-discriminant))) / (2*a)

    print("Roots are:", root1, "and", root2)

# Cubic
elif choice == 2:
    print("\nCubic Equation")

    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    coefficients = [a, b, c, d]
    roots = np.roots(coefficients)

    print("Roots are:")
    for r in roots:
        print(r)

# Quartic
elif choice == 3:
    print("\nQuartic Equation")

    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)

    print("Roots are:")
    for r in roots:
        print(r)

else:
    print("Invalid choice")
```
