```python
# Input from user
age = int(input("Enter staff age: "))
experience = int(input("Enter years of work experience: "))

# Determine Annual Tax Revenue
if experience > 25 and age >= 55:
    atrv = 5600000
elif experience > 20 and age >= 45:
    atrv = 4480000
elif experience > 10 and age >= 35:
    atrv = 1500000
else:
    atrv = 550000

# Output result
print("Annual Tax Revenue (ATRV): ₦", atrv)
```
