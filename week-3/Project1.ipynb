```python
import tkinter as tk
from tkinter import messagebox
import sqlite3


    
def verify_credentials(u, d):
    conn = sqlite3.connect('company_data.db')
    cursor = conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS employees (name TEXT, department TEXT)')
    cursor.execute('SELECT * FROM employees WHERE name = ? AND department = ?', (u, d))
    result = cursor.fetchone()
    conn.close()
    return result

def welcomeMessage(username):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x200")
    
    label_1 = tk.Label(window, text=f"Welcome {username}\n")
    label_1.pack()
    
    label_2 = tk.Label(window, text="This is Python GUI with Tkinter")
    label_2.pack()

def submit():
    user = username_entry.get()
    dept = department_entry.get()

    if verify_credentials(user, dept):
        welcomeMessage(user)
    else:
        messagebox.showinfo("Access Denied", f"Hello {user}, you are not a member of {dept}.")

root = tk.Tk()
root.title("Login")
root.geometry("500x500")

tk.Label(root, text="Name:").pack()
username_entry = tk.Entry(root)
username_entry.pack()

tk.Label(root, text="Department:").pack()
department_entry = tk.Entry(root)
department_entry.pack()

tk.Button(root, text="Login", command=submit).pack()

root.mainloop()
```
