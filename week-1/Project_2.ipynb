```python
# Compound Interest

P = float(input("Enter principle amount (P): "))
R = float(input("Enter annual interest rate (R %): ")) / 100
n = int(input("Enter number of times interest applied per year (n): "))
t = float(input("Enter number of years (t): "))

A = P * (1 + R / n) * (n * t)

print("Final Amount =", A)
```
